```python
# OBJETIVO: VOCÊ TRABALHA EM UMA EMPRESA QUE TEM UMA GRANDE BAE DE DADOS DOS CLIENTES. VOCÊ DEVE ENCONTRAR O MOTIVO DOS CANCELAMENTOS DE CONTRATOS RECENTES E PROPOR UMA SOLUÇÃO. TAMBÉM DEVE MOSTRAR AS ANÁLISES EM FORMA DE GRÁFICOS.

# 1) IMPORTAR A BASE DE DADOS

import pandas as pd

dados = pd.read_csv(r"C:\Users\mathy\OneDrive\Área de Trabalho\PROGRAMAÇÃO\Jornada Python\Python-tools\Análise de Dados\Arquivos auxiliares\cancelamentos.csv")   # importa a base de dados e armazena na variável 'dados'
```

```python
# 2) VISUALIZAR A BASE DE DADOS

display(dados)
```

```python
# 3) TRATAR A BASE DE DADOS

dados = dados.drop('CustomerID', axis = 1)   # DELETA A COLUNA 'CustomerID', POIS É APENAS UM CÓDIGO ALEATÓRIO DO CLIENTE E NÃO FAZ DIFERENÇA NA ANÁLISE. INFORMAÇÃO QUE NÃO AJUDA, ATRAPALHA. O ARGUMENTO 'axis' SERVE PARA ESPECIFICAR SE É UMA LINHA (axis = 0) OU UMA COLUNA (axis = 1)

print(dados.info())
```

```python
# VEJA QUE HÁ MAIS LINHAS TOTAIS DO QUE LINHAS NÃO-NULAS ('Non-Null Count'). ISSO PORQUE EXISTEM LINHAS COM ALGUMA INFORMAÇÃO EM BRANCO. VAMOS DELETAR ESSAS LINHAS UTILIZANDO O COMANDO:

dados = dados.dropna()   # DELETA AS LINHAS QUE CONTÉM PELO MENOS 1 INFORMAÇÃO VAZIA. O COMANDO '.dropna()' possui como parâmetros-padrão: 'dropna(axis = 0, any)'

print(dados.info())
```

```python
# 4) ANALIZAR COMO ESTÃO OS CANCELAMENTOS

# VAMOS ANALIZAR A TAXA DE CANCELAMENTO

print(dados['cancelou'].value_counts(normalize = True))   # MOSTRA A PORCENTAGEM DE CADA OPÇÃO DA COLUNA 'cancelou'. 1 significa sim e 0 significa não.
```

```python
# 5) ANALIZAR COM PROFUNDIDADE E ENCONTRAR O MOTIVO DO PROBLEMA E PROPOR UMA SOLUÇÃO

    # NÃO HÁ UM CAMINHO CERTO LOGO DE CARA, MAS DEVEMOS ESCOLHER UMA INFORMAÇÃO E VER SE ELA TEM RELAÇÃO COM OS CANCELAMENTOS. NESSE CASO, ANALIZAREMOS A RELAÇÃO DA COLUNA 'duracao_contrato' COM OS CANCELAMENTOS

print(dados['duracao_contrato'].value_counts(normalize = True))
```

```python
# 19% DOS CLIENTES POSSUEM/POSSUIAM O CONTRATO MENSAL.

numeric_df = dados.select_dtypes(include = 'number')   # CRIA UM DATAFRAME APENAS COM AS INFORMAÇÕES NUMÉRICAS

numeric_columns = []
for coluna in numeric_df.columns:   # CRIA UMA LISTA SOMENTE COM AS COLUNAS NUMÉRICAS
    numeric_columns.append(coluna)

display(dados.groupby('duracao_contrato')[numeric_columns].mean())   # MOSTRA O AGRUPAMENTO DA COLUNA 'duracao_contrato' EM RELAÇÃO ÀS COLUNAS NUMÉRICAS, CALCULANDO A MÉDIA DE CADA UMA DELAS

display(dados.groupby('duracao_contrato').count())
```
