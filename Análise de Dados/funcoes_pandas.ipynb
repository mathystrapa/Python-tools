```python
# O objetivo desse arquivo é explicar as funções da biblioteca pandas usadas em análises de dados


                                                # IMPORTAÇÃO DOS DADOS

# O primeiro passo para qualquer análise de dados é, obviamente, importar sua base de dados. Para isso, armazenamos a base de dados em uma variável qualquer da seguinte forma:

    #   dados = pandas.read_'extensão do arquivo'("caminho do arquivo")   /   o pandas reconhece várias extensões diferentes, que aparecem como sugestões ao programar no VS code

# Outros argumentos podem ser colocados após o caminho do arquivo. Em arquvios '.csv', podemos especificar o separador correto do arquivo. A biblioteca pandas utiliza como valor padrão 'sep = ',''. Porém, certifique-se que sua base de dados utilize ',' como separador. Em países onde a ',' é utilizada como separador decimal, como no Brasil, é comum encontrar bases de dados nesse modelo que utilizam ';' ou outro caractere como separador. Se for o caso, utilize 'sep = ';''.
# Em planilhas do Excel, podemos especificar a planilha específica que queremos importar. A biblioteca pandas utiliza como valor padrão 'sheet_name = 0', o que seleciona a primeira planilha do arquivo excel. Você também pode especificar a planilha pelo nome, como 'sheet_name = 'nome_da_planilha''. Esses são os 2 argumentos mais utilizados, mas existem outros argumentos para personalizar a importação na documentação oficial do panda: https://pandas.pydata.org/pandas-docs/stable/reference/io.html

# Exemplo:

import pandas

base_de_dados = pandas.read_csv(r"C:\Users\mathy\OneDrive\Área de Trabalho\PROGRAMAÇÃO\Jornada Python\Python-tools\Análise de Dados\Arquivos auxiliares\cancelamentos.csv", sep = ',')

display(base_de_dados)
```

```python
                                            # VISUALIZAÇÃO DA BASE DE DADOS

# '.head()': retorna as n primeiras linhas do DataFrame. Valor padrão: 5

display(base_de_dados.head())   # mostra as 5 primeiras linhas do DataFrame
display(base_de_dados.head(base_de_dados.shape[0]))   # mostra todas as linhas do DataFrame. Retorna o mesmo que 'display(base_de_dados)'
```

```python
# '.shape': Retorna uma tupla contendo a quantidade de linhas e colunas do DataFrame, respectivamente.

print('Linhas:', base_de_dados.shape[0])
print('Colunas:', base_de_dados.shape[1])
```
